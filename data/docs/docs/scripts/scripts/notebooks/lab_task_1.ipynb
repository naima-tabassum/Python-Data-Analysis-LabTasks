## 4. Pandas — Data Selection

You often need to pick specific rows or columns from your table. Pandas gives you two powerful tools:

| Tool | Stands for | Selects by |
|------|-----------|------------|
| `loc[]` | **loc**ation by label | Row/column **names** or **labels** |
| `iloc[]` | **i**nteger **loc**ation | Row/column **numbers** (0, 1, 2, ...) |


````python
# df.loc[row_label] — Select rows by their INDEX LABEL
# Here index labels are 0, 1, 2, 3, ... (the row numbers on the left)

# Select rows 1 to 4 (INCLUSIVE — both 1 and 4 are included)
df.loc[1:4]
````

````python
# df.iloc[row_position, column_position] — Select by POSITION NUMBER
# Remember: Python starts counting from 0!
# Slices are EXCLUSIVE at the end (like Python lists)

# Select rows 1,2,3,4 and columns from position 2 onwards
df.iloc[1:5, 2:]
````

````python
# Select a single column (returns a Series, not a DataFrame)
df['Temperature']
````

````python
# Select multiple columns (returns a DataFrame)
# Note the double brackets: [[ ]]
df[['EST', 'Temperature', 'Humidity']]
````

## 6. Pandas — Missing Values

Real-world data is almost never perfect. There will be **missing values** — represented as `NaN` (Not a Number).

You need to detect them and decide what to do with them.


````python
# df.isnull() — Check every cell: is it empty? Returns True/False
# True = missing (NaN), False = has a value
df.isnull().head(10)  # Show first 10 rows of the True/False table
````

````python
# df.isnull().sum() — Count the missing values in EACH column
# This is more useful than seeing a True/False table!
print('Missing value count per column:')
print(df.isnull().sum())
````

````python
# Total missing values in the entire DataFrame
total_missing = df.isnull().sum().sum()
print(f'Total missing values in the dataset: {total_missing}')
````

````python
# df.fillna(value) — Fill NaN with a specific value
# ⚠️ WARNING: Without inplace=True, the original df is NOT changed!
df_filled = df.fillna(0)  # Returns a new DataFrame, does NOT modify 'df'
print('After fillna (original df is unchanged):')
print(df_filled.isnull().sum())
````

````python
# To ACTUALLY change the original DataFrame, use inplace=True
# OR reassign: df = df.fillna(0)

# Fill numeric columns with their mean, text columns with 'None'
df_clean = df.copy()
df_clean['DewPoint'] = df_clean['DewPoint'].fillna(df_clean['DewPoint'].mean())
df_clean['Events'] = df_clean['Events'].fillna('No Event')

print('Missing values after smart fill:')
print(df_clean[['DewPoint', 'Events']].isnull().sum())
````

````python
# df.replace(old_value, new_value) — Replace a specific value with another
# Example: Replace all 0s with NaN
df_replaced = df.replace(0, np.nan)
print('Replace 0 with NaN done.')
````

````python
# df.interpolate() — Fill NaN by estimating from nearby values
# Good for time series data where values change gradually
# Example: if values are [10, NaN, 14], it fills NaN with 12 (average of 10 and 14)
df_interp = df[['Temperature', 'DewPoint']].interpolate()
print('After interpolation:')
print(df_interp.isnull().sum())
````

## 7. Pandas — Index Operations

The **index** is like the row label on the left side of your DataFrame.
By default it's 0, 1, 2, 3 ... but you can set any column as the index.

````python
# df.set_index('column') — Use a column as the row index
# Now you can use date values to look up rows with loc[]
df_indexed = df.set_index('EST')  # Returns new DataFrame; original 'df' is unchanged
print('New index is the date column:')
df_indexed.head()
````

````python
# Now you can select rows by date label
df_indexed.loc['1/1/2017':'1/4/2017']
````

````python
# df.reset_index() — Move the index BACK into a regular column
# ⚠️ Without inplace=True, the original df is not changed
df_reset = df_indexed.reset_index()
print('Index is reset to 0, 1, 2, ... and EST is back as a column:')
df_reset.head()
````

## 8. Pandas — Value Functions & Filtering

These are the functions that help you **analyze** and **filter** your data.

````python
# Statistical functions on a column
print(f'Max Temperature:  {df["Temperature"].max()}')
print(f'Min Temperature:  {df["Temperature"].min()}')
print(f'Mean Temperature: {float(df["Temperature"].mean()):.2f}')
print(f'Std Deviation:    {df["Temperature"].std():.2f}')  # How spread out the values are
print(f'Sum of all Temps: {df["Temperature"].sum()}')
````

````python
# Select first 5 rows of a specific column
df['EST'][0:5]
````

````python
# Boolean Masking — Filtering rows based on a condition
# Step 1: Create a boolean mask (True/False for each row)
mask = df['Temperature'] > 30
print('Boolean mask (True = Temperature > 30):')
print(mask.head(10))
````

````python
# Step 2: Apply the mask to filter the DataFrame
# This keeps only the rows where the condition is True
high_temp = df[df['Temperature'] > 30]
print(f'Rows where Temperature > 30: {len(high_temp)}')
high_temp[['EST', 'Temperature']].head()
````

````python
# Filter specific columns where a condition is met
# Show EST and Temperature columns where Temperature > 30
df[['EST', 'Temperature', 'VisibilityMiles']][df['Temperature'] > 30]
````

````python
# Filter rows where Events == 'Rain'
rain_days = df[df['Events'] == 'Rain']
print(f'Number of rainy days: {len(rain_days)}')
rain_days[['EST', 'Temperature', 'VisibilityMiles']]
````

````python
# Show all unique weather events
print('All unique weather events:')
print(df['Events'].unique())
````

## 9. NumPy — Array Creation

An **array** is like a list, but designed for math. All elements must be the same type.

````python
# Create a 1D array (like a simple list of numbers)
a = np.array([1, 2, 3, 4])
print('1D Array:', a)

# The MAGIC of NumPy: math operations apply to EVERY element at once!
# No need for loops!
print('a * 2:', a * 2)  # Output: [2 4 6 8]
print('a + 10:', a + 10)  # Output: [11 12 13 14]
````

````python
# Create a 2D array (like a table/matrix)
# Think of it as a list of lists
b = np.array([
    [1, 2],   # Row 0
    [3, 4],   # Row 1
    [5, 6]    # Row 2
])
print('2D Array:')
print(b)
````

````python
# np.zeros() — Create an array filled with all zeros
# Pass a list [rows, columns] to specify the shape
c = np.zeros([3, 4])  # 3 rows, 4 columns, all zeros
print('Zeros array:')
print(c)
````

````python
# np.ones() — Create an array filled with all ones
ones = np.ones([3, 4])
print('Ones array:')
print(ones)

# Multiply to fill with any number you want
nines = np.ones([3, 4]) * 9
print('\nNines array (ones * 9):')
print(nines)
````

````python
# np.full() — Create an array filled with a specific value directly
sevens = np.full((3, 3), 7)  # 3x3 array filled with 7
print('Full array (all 7s):')
print(sevens)
````

````python
# np.random.randn() — Create array with random numbers (from normal distribution)
# Useful for generating test data, simulations, ML weights
X = np.random.randn(4, 3)  # 4 rows, 3 columns, random values
print('Random array (4x3):')
print(X)
````

## 10. NumPy — Array Properties

After creating an array, you can inspect its properties.

````python
# .shape — Returns (rows, columns, ...)
# For 1D: (n,)  |  For 2D: (rows, cols)  |  For 3D: (layers, rows, cols)
print(f'Shape of a (1D): {a.shape}')  # Output: (4,)
print(f'Shape of b (2D): {b.shape}')  # Output: (3, 2)
print(f'Shape of c (2D): {c.shape}')  # Output: (3, 4)
````

````python
# .ndim — Number of dimensions
# 1D array → ndim = 1
# 2D array → ndim = 2
print(f'Dimensions of a: {a.ndim}')  # 1
print(f'Dimensions of b: {b.ndim}')  # 2
print(f'Dimensions of c: {c.ndim}')  # 2
````

````python
# .dtype — Data type of the elements inside the array
# Common types: int64, float64, bool, object
print(f'Data type of a: {a.dtype}')  # int64
print(f'Data type of c: {c.dtype}')  # float64 (zeros are floats by default)
````

## 11. NumPy — Range & Special Arrays

NumPy has convenient functions to create arrays without typing every number.

````python
# np.arange(start, stop, step) — Like Python's range(), but returns an array
# ⚠️ STOP is EXCLUSIVE (not included)

print('np.arange(0, 10, 3):', np.arange(0, 10, 3))   # 0, 3, 6, 9
print('np.arange(5):       ', np.arange(5))           # 0, 1, 2, 3, 4
print('np.arange(1, 11):   ', np.arange(1, 11))       # 1, 2, ..., 10
````

````python
# np.linspace(start, stop, num) — Create 'num' evenly spaced values
# ⚠️ STOP IS INCLUSIVE here (unlike arange)
# Great for creating smooth ranges for graphs and functions

print('np.linspace(1, 10, 6):', np.linspace(1, 10, 6))
# Gives: 1.0, 2.8, 4.6, 6.4, 8.2, 10.0 — 6 values evenly spaced from 1 to 10
````

## 12. NumPy — Slicing & Reshaping

You can pick specific rows and columns from a NumPy array, similar to how you use `iloc[]` in Pandas.

````python
# .reshape() — Change the shape of an array without changing its data
# Total elements must stay the same: 12 elements → can reshape to (3,4) or (4,3) or (2,6) etc.

arr = np.arange(1, 13).reshape(3, 4)
print('Reshaped array (3 rows, 4 columns):')
print(arr)
````

````python
# 2D Array Slicing: array[row_start:row_end, col_start:col_end]
# This is the same slice notation as Python lists, but for 2D!

x = np.array([
    [1,  2,  3,  55, 56, 57],  # Row 0
    [4,  5,  6,  66, 67, 68],  # Row 1
    [7,  8,  9,  99, 96, 97],  # Row 2
    [33, 44, 50, 23, 45, 23]   # Row 3
])

print('Full array:')
print(x)
````

````python
# Select rows 1 and 2, columns 1 and 2
# x[row_start:row_end, col_start:col_end]
# End is EXCLUSIVE in slicing

selected = x[1:3, 1:3]
print('x[1:3, 1:3] selects:')
print(selected)
#   Row 1, Col 1 = 5  |  Row 1, Col 2 = 6
#   Row 2, Col 1 = 8  |  Row 2, Col 2 = 9
````

````python
# More slicing examples
print('First row:        ', x[0, :])     # All columns of row 0
print('Last column:      ', x[:, -1])    # All rows, last column
print('Bottom-right 2x2:')
print(x[2:, 4:])                         # Last 2 rows, last 2 columns
````

## 13. NumPy — Axis Operations

This is one of the **most confusing** concepts for beginners. Let's break it down clearly.

```
axis=0  →  Operate DOWN the columns (across rows)   → gives 1 result per COLUMN
axis=1  →  Operate ACROSS the rows  (across columns) → gives 1 result per ROW
```

**Memory trick:** axis=0 collapses rows. axis=1 collapses columns.

````python
# In data science: rows = samples, columns = features
X = np.array([
    [10, 20, 30],  # Sample 1: Feature A=10, Feature B=20, Feature C=30
    [40, 50, 60],  # Sample 2: Feature A=40, Feature B=50, Feature C=60
    [70, 80, 90],  # Sample 3: Feature A=70, Feature B=80, Feature C=90
    [10, 30, 50],  # Sample 4: Feature A=10, Feature B=30, Feature C=50
])
print('X shape:', X.shape)  # (4 samples, 3 features)
````

````python
# axis=0 — Compute COLUMN-WISE (feature-wise)
# "What is the average value of each FEATURE across all samples?"
col_mean = X.mean(axis=0)
print('Mean of each column (axis=0):', col_mean)
# Result has 3 values — one per column/feature
````

````python
# axis=1 — Compute ROW-WISE (sample-wise)
# "What is the total of each SAMPLE across all features?"
row_sum = X.sum(axis=1)
print('Sum of each row (axis=1):', row_sum)
# Result has 4 values — one per row/sample
````

````python
# Global operations (no axis = whole array)
print(f'Global sum:  {X.sum()}')
print(f'Global mean: {X.mean()}')
print(f'Global max:  {X.max()}')
print(f'Global min:  {X.min()}')
````
